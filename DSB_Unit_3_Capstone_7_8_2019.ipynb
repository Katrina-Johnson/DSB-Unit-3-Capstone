```python
#Random Forest Model

from sklearn.ensemble import RandomForestRegressor

tree_regr = RandomForestRegressor(max_depth=50, n_estimators=3000) #Do not use n_jobs=-1 on my own comput4er, because mycomputer needs at least one CPU core for other things.
tree_regr.fit(X_train, y_train)  
#RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=2,
           #max_features='auto', max_leaf_nodes=None,
           #min_impurity_decrease=0.0, min_impurity_split=None,
           #min_samples_leaf=1, min_samples_split=2,
           #min_weight_fraction_leaf=0.0, n_estimators=100, n_jobs=None,
           #oob_score=False, random_state=0, verbose=0, warm_start=False)
          
# Default behavior: max_depth = None. (This usually results in a lot of overfitting.) The only limiting behavior is: min_sample_split=2.
          
importances = pd.Series(tree_regr.feature_importances_, index=X_train.columns)          

importances = 100.0 * (importances / importances.max())

print(importances.sort_values(ascending=False))


from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_regr, X_train, y_train, cv=3)
print(scores) 

print("Accuracy: %0.5f (+/- %0.5f)" % (scores.mean(), scores.std() * 2))

print(tree_regr.score(X_train, y_train))
print(tree_regr.score(X_test, y_test))

y_pred = tree_regr.predict(X_test)

residual = y_pred - y_test

#A skadasticity scatter plot is equivilant to a confusion matrix, and can be used for regression modeling.  See lesson 2.4.3

plt.scatter(y_pred, residual) #This measures the skedasticity, which is the relationship between the prediction and what the actual number of error is.  
plt.xlabel('Residual')
plt.ylabel('Random Forest Scedasticity')
plt.show()

# Remember the residual y_test - y_pred
```

```python
from sklearn import ensemble 

params = {'n_estimators': 2000, 'max_depth': 5, 'min_samples_split': 2,
          'learning_rate': 0.15, 'loss': 'ls'} 

#Default n_estimators 100, default max_depth is 3.  The idea is that the next tree will fix the problems from the last tree.  
# It assumes that each tree will have a lot of errors.

#The learning_rate is the rate at which the model uses the incorrect estimates from previous trees to make the current tree.
#The learning rate shrinks the contribution of the next tree by that amount.  There is a trade off between the learning rate and the number of estimators.
# The default learning rate is .1;  this makes it so that leach tree has a tenth of the importance of the tree before it.  
#I can experiment with changing the learning rate from .1 to .01.
#I can also experiment with the max_depth.  Take it to 30, and see what happens.
#I am having good result with 1000 estimators with random forest.  
#Having lots of estimators does not tend to greatly increase overfitting in gradient boost models.
#I can try using more estimators, in order to increase accuracy in the gradient boosting model.

gbr = ensemble.GradientBoostingRegressor(**params)

gbr.fit(X_train, y_train)


importances = pd.Series(gbr.feature_importances_, index=X_train.columns)          

importances = 100.0 * (importances / importances.max())

print(importances.sort_values(ascending=False))


scores = cross_val_score(gbr, X_train, y_train, cv=3)
print(scores) 

print("Accuracy: %0.5f (+/- %0.5f)" % (scores.mean(), scores.std() * 2))

print(gbr.score(X_train, y_train))
print(gbr.score(X_test, y_test))

y_pred = gbr.predict(X_test)

residual = y_pred - y_test

#A skadasticity scatter plot is equivilant to a confusion matrix, and can be used for regression modeling.  See lesson 2.4.3

plt.scatter(y_pred, residual) #This measures the skedasticity, which is the relationship between the prediction and what the actual number of error is.  
plt.xlabel('Residual')
plt.ylabel('y_pred')
plt.title('Gradient Boosting Scedasticity')
plt.show()

# Remember the residual y_test - y_pred
```
